# Load Data

This file was also included with the dataset and (also?) permit to load the dataset and display one of each class.

It has been modified in order to remove skimage dependency (but not sure if it works well because all images are classified as 0).

```python
import glob, os 
# from skimage import io, transform
import numpy as np
import matplotlib.pyplot as plt
```

```python
path="images/"
imlist= glob.glob(os.path.join(path, '*.jpg'))
```

```python
def dataset(file_list,size=(300,180),flattened=False):
    data = []
    for i, file in enumerate(file_list):
        image = plt.imread(file)
        #image = io.imread(file)
        #image = transform.resize(image, size, mode='constant')
        if flattened:
            image = image.flatten()

        data.append(image)

    labels = [1 if f.split("/")[-1][0] == 'P' else 0 for f in file_list]

    return np.array(data), np.array(labels)
```

```python
# Load the dataset (may take a few seconds)
X,y=dataset(imlist)
```

```python
# X has the following structure: X[imageid, y,x,channel]
print('X: ',X.shape)  # data
print('y: ',y.shape)  # target
```

```python
print('Class 0: ',sum(y==0))
print('Class 1: ',sum(y==1))
print('Total  : ',len(y))
print(y)
```
